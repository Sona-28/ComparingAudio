```python
import librosa
ad1 = "C:\\Users\\sonas\\Music\\cat_meow.wav"
ad2 = "C:\\Users\\sonas\\Music\\Cat_Meow_2.wav"
y1, sr1 = librosa.load(ad1)
y2, sr2 = librosa.load(ad2)

```

```python
%matplotlib inline
import sklearn
import matplotlib.pyplot as plt
import librosa.display
```

```python
spectral_centroids1 = librosa.feature.spectral_centroid(y1, sr=sr1)[0]
spectral_centroids2 = librosa.feature.spectral_centroid(y2, sr=sr2)[0]
a = spectral_centroids1.shape
b = spectral_centroids2.shape
```

```python
if(a==b):
    print("Match")
else:
    print("Unmatch")
```

```python
# MFCC
mfccs1 = librosa.feature.mfcc(y1, sr=sr1)
mfccs2 = librosa.feature.mfcc(y2, sr=sr2)
m1= mfccs1.shape
m2=mfccs2.shape
```

```python
if(m1==m2):
    print("Matched audio")
else:
    print("Unmatched audio")
```

```python
librosa.display.specshow(mfccs1, sr=sr1, x_axis='time')
```

```python
librosa.display.specshow(mfccs2, sr=sr2, x_axis='time')
```

```python
librosa.display.waveshow(y1, sr=sr1)
librosa.display.waveshow(y2, sr=sr2)
```

```python
# Computing the time variable for visualization
plt.figure(figsize=(20,5))
frames1 = range(len(spectral_centroids1))
t1 = librosa.frames_to_time(frames1)
frames2= range(len(spectral_centroids2))
t2 = librosa.frames_to_time(frames2)
# Normalising the spectral centroid for visualisation
def normalize(y1, axis=0):
    return sklearn.preprocessing.minmax_scale(y1, axis=axis)
def normalize(y2, axis=0):
    return sklearn.preprocessing.minmax_scale(y2, axis=axis)

#Plotting the Spectral Centroid along the waveform
w1=librosa.display.waveshow(y1, sr=sr1, alpha=0.4)
plt.plot(t1, normalize(spectral_centroids1), color='r')
librosa.display.waveshow(y2, sr=sr2, alpha=0.4)
w2=plt.plot(t2, normalize(spectral_centroids2), color='g')
```
